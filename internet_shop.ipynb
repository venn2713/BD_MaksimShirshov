```python
%%sql
DROP DATABASE IF EXISTS internet_shop;
CREATE DATABASE internet_shop DEFAULT CHARACTER SET utf8mb4;
USE internet_shop;

CREATE TABLE customers (
    customer_id INT AUTO_INCREMENT,
    last_name VARCHAR(100) NOT NULL,
    first_name VARCHAR(100) NOT NULL,
    middle_name VARCHAR(100),
    address VARCHAR(200) NOT NULL,
    phone VARCHAR(15),
    email VARCHAR(50) NOT NULL,
    regular_customer ENUM('0', '1') DEFAULT '0',
    discount DECIMAL(5, 2),
    PRIMARY KEY (customer_id)
) ENGINE=InnoDB;

CREATE TABLE products (
    product_id INT AUTO_INCREMENT,
    name VARCHAR(100) NOT NULL,
    price DECIMAL(10, 2) NOT NULL,
    unit VARCHAR(20) NOT NULL,
    PRIMARY KEY (product_id)
) ENGINE=InnoDB;

CREATE TABLE sales (
    sale_id INT AUTO_INCREMENT,
    product_id INT NOT NULL,
    customer_id INT NOT NULL,
    sale_date TIMESTAMP DEFAULT CURRENT_TIMESTAMP,
    delivery_date TIMESTAMP DEFAULT CURRENT_TIMESTAMP,
    quantity INT NOT NULL,
    discount DECIMAL(5, 2),
    PRIMARY KEY (sale_id),
    FOREIGN KEY (product_id) REFERENCES products(product_id) ON DELETE CASCADE ON UPDATE CASCADE,
    FOREIGN KEY (customer_id) REFERENCES customers(customer_id) ON DELETE CASCADE ON UPDATE CASCADE
) ENGINE=InnoDB;

CREATE TABLE warehouse (
    product_id INT NOT NULL,
    quantity_in_stock INT NOT NULL,
    PRIMARY KEY (product_id),
    FOREIGN KEY (product_id) REFERENCES products(product_id) ON DELETE CASCADE ON UPDATE CASCADE
) ENGINE=InnoDB;

CREATE TABLE order_details (
    order_detail_id INT AUTO_INCREMENT,
    sale_id INT NOT NULL,
    product_id INT NOT NULL,
    quantity INT NOT NULL,
    discount DECIMAL(5, 2),
    PRIMARY KEY (order_detail_id),
    FOREIGN KEY (sale_id) REFERENCES sales(sale_id) ON DELETE CASCADE ON UPDATE CASCADE,
    FOREIGN KEY (product_id) REFERENCES products(product_id) ON DELETE CASCADE ON UPDATE CASCADE
) ENGINE=InnoDB;

-- Данные таблицы Customers
INSERT INTO customers (last_name, first_name, middle_name, address, phone, email, regular_customer, discount) 
VALUES 
('Smith', 'John', 'Anthony', '123 Main St', '555-1234', 'johnsmith@example.com', '1', 5.00),
('Doe', 'Jane', 'Elizabeth', '456 Maple St', '555-5678', 'janedoe@example.com', '0', 0.00),
('Johnson', 'Mike', 'Edward', '789 Pine St', '555-2468', 'mikejohnson@example.com', '1', 10.00),
('Brown', 'Emily', 'Grace', '1600 Amphitheatre Parkway', '555-3344', 'emilybrown@example.com', '1', 3.00),
('Taylor', 'Robert', 'James', '1 Infinite Loop', '555-7788', 'roberttaylor@example.com', '0', 0.00),
('Wilson', 'Emma', 'Sophia', '1 Hacker Way', '555-1122', 'emmawilson@example.com', '1', 15.00),
('Miller', 'Charlie', 'Daniel', '1601 Willow Rd', '555-9988', 'charliemiller@example.com', '0', 0.00),
('Davis', 'Olivia', 'Mia', '3555 Monarch St', '555-6677', 'oliviadavis@example.com', '1', 7.00),
('Moore', 'Lucas', 'Matthew', '2701 San Tomas Expressway', '555-4422', 'lucasmoore@example.com', '0', 0.00),
('Clark', 'Ava', 'Isabella', '3000 Tannery Way', '555-8877', 'avaclark@example.com', '1', 5.00);

-- Данные таблицы Products
INSERT INTO products (name, price, unit) 
VALUES 
('Apple', 0.99, 'kg'),
('Orange', 1.49, 'kg'),
('Milk', 2.50, 'Litre'),
('Banana', 0.59, 'kg'),
('Cheese', 4.99, 'kg'),
('Bread', 1.20, 'Piece'),
('Tomato', 1.99, 'kg'),
('Eggs', 2.10, 'Dozen'),
('Chicken', 5.99, 'kg'),
('Butter', 2.30, 'Block');

-- Данные таблицы Sales
INSERT INTO sales (product_id, customer_id, sale_date, delivery_date, quantity, discount) 
VALUES 
(1, 1, NOW(), NOW() + INTERVAL 1 DAY, 10, 5.00),
(2, 2, NOW(), NOW() + INTERVAL 2 DAY, 15, 0.00),
(3, 3, NOW(), NOW() + INTERVAL 3 DAY, 5, 10.00),
(4, 4, NOW(), NOW() + INTERVAL 2 DAY, 5, 3.00),
(5, 5, NOW(), NOW() + INTERVAL 3 DAY, 2, 0.00),
(6, 6, NOW(), NOW() + INTERVAL 1 DAY, 10, 15.00),
(7, 7, NOW(), NOW() + INTERVAL 2 DAY, 3, 0.00),
(8, 1, NOW(), NOW() + INTERVAL 1 DAY, 7, 7.00),
(9, 2, NOW(), NOW() + INTERVAL 1 DAY, 6, 0.00),
(10, 3, NOW(), NOW() + INTERVAL 1 DAY, 4, 5.00);

-- Данные таблицы Warehouse
INSERT INTO warehouse (product_id, quantity_in_stock) 
VALUES 
(1, 100),
(2, 200),
(3, 300),
(4, 500),
(5, 150),
(6, 120),
(7, 80),
(8, 220),
(9, 300),
(10, 250);

-- Данные таблицы Order Details
INSERT INTO order_details (sale_id, product_id, quantity, discount) 
VALUES 
(1, 1, 10, 5.00),
(2, 2, 15, 0.00),
(3, 3, 5, 10.00),
(4, 4, 5, 3.00),
(5, 5, 2, 0.00),
(6, 6, 10, 15.00),
(7, 7, 3, 0.00),
(8, 1, 7, 7.00),
(9, 2, 6, 0.00),
(10, 3, 4, 5.00);
```

```python
import pandas as pd
from sqlalchemy import create_engine

# Создайте строку подключения к базе данных
connection_string = 'mysql+mysqlconnector://root:VoLk_2713@localhost/internet_shop'

# Создайте объект Engine с использованием строки подключения
connection = create_engine(connection_string)
```

```python
# Примеры запросов
# Простой выбор данных
query = "SELECT customer_id, last_name, first_name FROM customers"
result = pd.read_sql(query, connection)
result
```

```python
# Выборка нескольких полей из одной таблицы
query = "SELECT * FROM customers"
result = pd.read_sql(query, connection)
result
```

```python
# Выборка данных из двух связанных таблиц
query = """
SELECT customers.last_name, products.name
FROM customers, sales, products
WHERE customers.customer_id = sales.customer_id AND products.product_id = sales.product_id
"""
result = pd.read_sql(query, connection)
result
```

```python
# Выборка данных, сортировка по нескольким полям
query = "SELECT * FROM customers ORDER BY last_name, first_name"
result = pd.read_sql(query, connection)
result
```

```python
# Условия отбора
query = "SELECT * FROM customers WHERE last_name = 'Smith' AND first_name = 'John'"
result = pd.read_sql(query, connection)
result

query = "SELECT * FROM customers WHERE last_name = 'Smith' OR last_name = 'Johnson'"
result = pd.read_sql(query, connection)
result
```

```python
# Использование DISTINCT
query = "SELECT DISTINCT last_name FROM customers"
result = pd.read_sql(query, connection)
result
```

```python
# Внешнее объединение таблиц
query = """
SELECT customers.last_name, sales.sale_id
FROM customers LEFT JOIN sales ON customers.customer_id = sales.customer_id
"""
result = pd.read_sql(query, connection)
result
```

```python
# Использование предложений IS NULL, IS NOT NULL
query = "SELECT * FROM customers WHERE phone IS NULL"
result = pd.read_sql(query, connection)
result

query = "SELECT * FROM customers WHERE phone IS NOT NULL"
result = pd.read_sql(query, connection)
result
```

```python
# Использование LIKE
query = "SELECT * FROM customers WHERE last_name LIKE 'S%'"
result = pd.read_sql(query, connection)
result
```

```python
# Использование UNION
query = """
SELECT last_name FROM customers
UNION
SELECT name FROM products
"""
result = pd.read_sql(query, connection)
result
```

```python
# Группировка данных
# Создание итогового запроса
query = """
SELECT customers.last_name, COUNT(sales.sale_id) AS NumberOfSales
FROM customers JOIN sales ON customers.customer_id = sales.customer_id
GROUP BY customers.last_name
"""
result = pd.read_sql(query, connection)
result
```

```python
# Простой группирующий запрос
query = "SELECT product_id, COUNT(*) AS Count FROM sales GROUP BY product_id"
result = pd.read_sql(query, connection)
result
```

```python
# Группировка по нескольким полям
query = "SELECT customer_id, product_id, COUNT(*) AS Count FROM sales GROUP BY customer_id, product_id"
result = pd.read_sql(query, connection)
result
```

```python
# Группировка с предварительными вычислениями
query = """
SELECT customer_id, COUNT(*) AS Count, AVG(quantity) AS AverageQuantity
FROM sales GROUP BY customer_id HAVING AverageQuantity > 2
"""
result = pd.read_sql(query, connection)
result
```

```python
# Группировка с использованием вычисляемых полей
query = """
SELECT customer_id, COUNT(*) AS Count, AVG(quantity) AS AverageQuantity, SUM(price*quantity) AS TotalSales
FROM sales JOIN products ON sales.product_id = products.product_id
GROUP BY customer_id
"""
result = pd.read_sql(query, connection)
result
```

```python
# Подзапросы
# Запрос с использованием IN
query = """
SELECT * FROM customers WHERE customer_id IN (SELECT customer_id FROM sales WHERE product_id = 1)
"""
result = pd.read_sql(query, connection)
result
```

```python
# Использование ALL, ANY, EXISTS
query = """
SELECT product_id FROM sales WHERE quantity > ALL (SELECT AVG(quantity) FROM sales GROUP BY product_id)
"""
result = pd.read_sql(query, connection)
result
```

```python
query = """
SELECT product_id FROM sales WHERE quantity > ANY (SELECT AVG(quantity) FROM sales GROUP BY product_id)
"""
result = pd.read_sql(query, connection)
result
```

```python
query = """
SELECT * FROM customers WHERE EXISTS (SELECT * FROM sales WHERE sales.customer_id = customers.customer_id)
"""
result = pd.read_sql(query, connection)
result
```
