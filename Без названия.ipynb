```python
%%sql
DROP DATABASE IF EXISTS internet_shop;
CREATE DATABASE internet_shop DEFAULT CHARACTER SET utf8mb4;
USE internet_shop;

CREATE TABLE customers (
    customer_id INT AUTO_INCREMENT,
    last_name VARCHAR(100) NOT NULL,
    first_name VARCHAR(100) NOT NULL,
    middle_name VARCHAR(100),
    address VARCHAR(200) NOT NULL,
    phone VARCHAR(15),
    email VARCHAR(50) NOT NULL,
    regular_customer ENUM('0', '1') DEFAULT '0',
    discount DECIMAL(5, 2),
    PRIMARY KEY (customer_id)
) ENGINE=InnoDB;

CREATE TABLE products (
    product_id INT AUTO_INCREMENT,
    name VARCHAR(100) NOT NULL,
    price DECIMAL(10, 2) NOT NULL,
    unit VARCHAR(20) NOT NULL,
    PRIMARY KEY (product_id)
) ENGINE=InnoDB;

CREATE TABLE sales (
    sale_id INT AUTO_INCREMENT,
    product_id INT NOT NULL,
    customer_id INT NOT NULL,
    sale_date TIMESTAMP DEFAULT CURRENT_TIMESTAMP,
    delivery_date TIMESTAMP DEFAULT CURRENT_TIMESTAMP,
    quantity INT NOT NULL,
    discount DECIMAL(5, 2),
    PRIMARY KEY (sale_id),
    FOREIGN KEY (product_id) REFERENCES products(product_id) ON DELETE CASCADE ON UPDATE CASCADE,
    FOREIGN KEY (customer_id) REFERENCES customers(customer_id) ON DELETE CASCADE ON UPDATE CASCADE
) ENGINE=InnoDB;

CREATE TABLE warehouse (
    product_id INT NOT NULL,
    quantity_in_stock INT NOT NULL,
    PRIMARY KEY (product_id),
    FOREIGN KEY (product_id) REFERENCES products(product_id) ON DELETE CASCADE ON UPDATE CASCADE
) ENGINE=InnoDB;

CREATE TABLE order_details (
    order_detail_id INT AUTO_INCREMENT,
    sale_id INT NOT NULL,
    product_id INT NOT NULL,
    quantity INT NOT NULL,
    discount DECIMAL(5, 2),
    PRIMARY KEY (order_detail_id),
    FOREIGN KEY (sale_id) REFERENCES sales(sale_id) ON DELETE CASCADE ON UPDATE CASCADE,
    FOREIGN KEY (product_id) REFERENCES products(product_id) ON DELETE CASCADE ON UPDATE CASCADE
) ENGINE=InnoDB;

-- Данные таблицы Customers
INSERT INTO customers (last_name, first_name, middle_name, address, phone, email, regular_customer, discount) 
VALUES 
('Smith', 'John', 'Anthony', '123 Main St', '555-1234', 'johnsmith@example.com', '1', 5.00),
('Doe', 'Jane', 'Elizabeth', '456 Maple St', '555-5678', 'janedoe@example.com', '0', 0.00),
('Johnson', 'Mike', 'Edward', '789 Pine St', '555-2468', 'mikejohnson@example.com', '1', 10.00),
('Brown', 'Emily', 'Grace', '1600 Amphitheatre Parkway', '555-3344', 'emilybrown@example.com', '1', 3.00),
('Taylor', 'Robert', 'James', '1 Infinite Loop', '555-7788', 'roberttaylor@example.com', '0', 0.00),
('Wilson', 'Emma', 'Sophia', '1 Hacker Way', '555-1122', 'emmawilson@example.com', '1', 15.00),
('Miller', 'Charlie', 'Daniel', '1601 Willow Rd', '555-9988', 'charliemiller@example.com', '0', 0.00),
('Davis', 'Olivia', 'Mia', '3555 Monarch St', '555-6677', 'oliviadavis@example.com', '1', 7.00),
('Moore', 'Lucas', 'Matthew', '2701 San Tomas Expressway', '555-4422', 'lucasmoore@example.com', '0', 0.00),
('Clark', 'Ava', 'Isabella', '3000 Tannery Way', '555-8877', 'avaclark@example.com', '1', 5.00);

-- Данные таблицы Products
INSERT INTO products (name, price, unit) 
VALUES 
('Apple', 0.99, 'kg'),
('Orange', 1.49, 'kg'),
('Milk', 2.50, 'Litre'),
('Banana', 0.59, 'kg'),
('Cheese', 4.99, 'kg'),
('Bread', 1.20, 'Piece'),
('Tomato', 1.99, 'kg'),
('Eggs', 2.10, 'Dozen'),
('Chicken', 5.99, 'kg'),
('Butter', 2.30, 'Block');

-- Данные таблицы Sales
INSERT INTO sales (product_id, customer_id, sale_date, delivery_date, quantity, discount) 
VALUES 
(1, 1, NOW(), NOW() + INTERVAL 1 DAY, 10, 5.00),
(2, 2, NOW(), NOW() + INTERVAL 2 DAY, 15, 0.00),
(3, 3, NOW(), NOW() + INTERVAL 3 DAY, 5, 10.00),
(4, 4, NOW(), NOW() + INTERVAL 2 DAY, 5, 3.00),
(5, 5, NOW(), NOW() + INTERVAL 3 DAY, 2, 0.00),
(6, 6, NOW(), NOW() + INTERVAL 1 DAY, 10, 15.00),
(7, 7, NOW(), NOW() + INTERVAL 2 DAY, 3, 0.00),
(8, 1, NOW(), NOW() + INTERVAL 1 DAY, 7, 7.00),
(9, 2, NOW(), NOW() + INTERVAL 1 DAY, 6, 0.00),
(10, 3, NOW(), NOW() + INTERVAL 1 DAY, 4, 5.00);

-- Данные таблицы Warehouse
INSERT INTO warehouse (product_id, quantity_in_stock) 
VALUES 
(1, 100),
(2, 200),
(3, 300),
(4, 500),
(5, 150),
(6, 120),
(7, 80),
(8, 220),
(9, 300),
(10, 250);

-- Данные таблицы Order Details
INSERT INTO order_details (sale_id, product_id, quantity, discount) 
VALUES 
(1, 1, 10, 5.00),
(2, 2, 15, 0.00),
(3, 3, 5, 10.00),
(4, 4, 5, 3.00),
(5, 5, 2, 0.00),
(6, 6, 10, 15.00),
(7, 7, 3, 0.00),
(8, 1, 7, 7.00),
(9, 2, 6, 0.00),
(10, 3, 4, 5.00);
```
